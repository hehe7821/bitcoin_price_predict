```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# ===== 1) CSV 불러오기 =====
csv_path = "btc_1d_data_2018_to_2025.csv"  # <- 파일 경로
df = pd.read_csv(csv_path)

# 공백/개행 제거해서 컬럼명 정리
df.columns = [c.strip() for c in df.columns]

# (선택) 'Open time'이 있다면 쓰기 좋게 변환해두기 (밀리초 epoch일 가능성)
if 'Open time' in df.columns:
    try:
        # 숫자형 & 값이 큰 경우 밀리초 epoch로 가정
        if np.issubdtype(df['Open time'].dtype, np.number) and df['Open time'].max() > 1e11:
            df['Open time'] = pd.to_datetime(df['Open time'], unit='ms', errors='coerce')
        else:
            df['Open time'] = pd.to_datetime(df['Open time'], errors='coerce')
    except Exception:
        pass  # 변환 실패해도 상관없음(어차피 숫자 컬럼만 상관계산)

# ===== 2) 숫자형 컬럼만 선택 (시간/문자 제외) =====
num_df = df.select_dtypes(include=[np.number]).copy()

# NaN 처리(필요시)
num_df = num_df.replace([np.inf, -np.inf], np.nan).dropna(how="any")

# ===== 3) 상관행렬 계산 =====
corr_pearson = num_df.corr(method='pearson')
corr_spearman = num_df.corr(method='spearman')

# 보기 좋게 출력
print("=== Pearson correlation (rounded) ===")
print(corr_pearson.round(3))
print("\n=== Spearman correlation (rounded) ===")
print(corr_spearman.round(3))

# 파일로 저장
out_dir = Path("correlation_outputs")
out_dir.mkdir(exist_ok=True)
corr_pearson.to_csv(out_dir / "corr_pearson.csv", float_format="%.6f", encoding="utf-8-sig")
corr_spearman.to_csv(out_dir / "corr_spearman.csv", float_format="%.6f", encoding="utf-8-sig")
print(f"\nSaved to: {out_dir/'corr_pearson.csv'} / {out_dir/'corr_spearman.csv'}")

# ===== 4) 히트맵 (matplotlib) =====
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr_pearson.values, vmin=-1, vmax=1, cmap="coolwarm")
ax.set_xticks(range(len(corr_pearson.columns)))
ax.set_yticks(range(len(corr_pearson.index)))
ax.set_xticklabels(corr_pearson.columns, rotation=45, ha="right")
ax.set_yticklabels(corr_pearson.index)
ax.set_title("Pearson Correlation")
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()

# ===== 5) (선택) 상관 높은 상위 쌍만 추출 =====
# 위삼각만 뽑아 평탄화 → 절대값으로 정렬
mask = np.triu(np.ones_like(corr_pearson, dtype=bool), k=1)
pairs = (
    corr_pearson.where(mask)
    .stack()
    .rename("corr")
    .abs()
    .sort_values(ascending=False)
)
print("\nTop correlated pairs (|corr|):")
print(pairs.head(10))
```

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt

df = pd.read_csv("btc_1d_data_2018_to_2025.csv")
df.columns = [c.strip() for c in df.columns]
df = df.drop(columns=["Ignore"], errors="ignore")

# 1) 파생 특징 만들기(중복 줄이기)
df["hl_range"] = (df["High"] - df["Low"]) / df["Close"]
df["buy_ratio"] = df["Taker buy base asset volume"] / df["Volume"]
df["avg_trade_size"] = df["Quote asset volume"] / df["Number of trades"]
df["quote_per_base"] = df["Quote asset volume"] / df["Volume"]

# 2) 타깃: 다음날 로그수익률
df["ret"] = np.log(df["Close"]).diff()
df["ret_fwd1"] = df["ret"].shift(-1)

# 3) 후보 특징 선택
feat_cols = [
    "Close","hl_range","Volume","buy_ratio","avg_trade_size","quote_per_base",
    "Number of trades","Taker buy base asset volume"
]

# 4) 시차 상관표 만들기
LAGS = [1,2,3,5,10]
corr_mat = pd.DataFrame(index=feat_cols, columns=LAGS, dtype=float)

for col in feat_cols:
    for L in LAGS:
        corr_mat.loc[col, L] = df[col].shift(L).corr(df["ret_fwd1"])

print("Lagged corr with next-day return:")
print(corr_mat.round(3))

# 5) 히트맵
fig, ax = plt.subplots(figsize=(7,5))
im = ax.imshow(corr_mat.values.astype(float), vmin=-0.2, vmax=0.2, cmap="coolwarm")
ax.set_xticks(range(len(LAGS))); ax.set_xticklabels(LAGS)
ax.set_yticks(range(len(feat_cols))); ax.set_yticklabels(feat_cols)
ax.set_title("Corr(F_{t−lag}, ret_{t})")
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout(); plt.show()
```
