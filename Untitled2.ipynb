```python
df_gold = pd.read_csv('XAU_USD 과거 데이터.csv')
```

```python
# 예시 입력: '2025-08-24 00:00:00+00:00'
df['Open time'] = pd.to_datetime(df['Open time'], utc=True, errors='coerce')
# 날짜(UTC 자정으로 정규화)
df['date_utc'] = df['Open time'].dt.normalize()  # = yyyy-mm-dd 00:00:00+00:00
```

```python
df
```

```python
df_gold['날짜'] = df_gold['날짜'].astype(str).str.replace(' ', '', regex=False)
gold_date = pd.to_datetime(df_gold['날짜'], errors='coerce', format='%Y-%m-%d')
df_gold['date_utc'] = gold_date.dt.tz_localize('UTC').dt.normalize()
```

```python
df_gold
```

```python
df_gold['종가'] = (
    df_gold['종가']
    .astype(str)
    .str.replace(',', '', regex=False)
)
df_gold['Gold_close'] = pd.to_numeric(df_gold['종가'], errors='coerce')
```

```python
df_gold.tail()
```

```python
df_gold = df_gold[['date_utc', 'Gold_close']]
```

```python
merged = pd.merge(df, df_gold, on='date_utc', how='left')
merged = merged.sort_values('Open time').drop_duplicates(subset=['Open time'])
# merged = merged.drop(columns=['date_utc'])
```

```python
merged.to_csv('merged_btc_gold_2.csv', index=False, encoding='utf-8-sig')
print("병합 완료! -> merged_btc_gold_2.csv")
```
